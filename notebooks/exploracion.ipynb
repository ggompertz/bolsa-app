```python
# Probar con acción chilena
df_cl = get_ohlcv('COPEC.SN', interval='1d', period='1y')
print('Chile:', df_cl.tail(3))

# Probar con acción USA
df_us = get_ohlcv('AAPL', interval='1d', period='1y')
print('USA:', df_us.tail(3))
```

```python
# Calcular todos los indicadores
df = add_all_indicators(df_us.copy())
print(df.columns.tolist())
df.tail(5)
```

```python
# Gráfico de velas completo
chart_json = build_candlestick_chart(
    df, symbol='AAPL',
    indicators=['SMA_20', 'SMA_50', 'Volume', 'RSI', 'MACD'],
    support_resistance=find_support_resistance(df),
    patterns=get_all_patterns(df),
    show_volume=True,
)

import plotly.io as pio
fig = pio.from_json(chart_json)
fig.show()
```
